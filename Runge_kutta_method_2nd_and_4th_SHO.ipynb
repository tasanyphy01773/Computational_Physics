```python
import math 
import scipy
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# initial condition
x = [1] #position
v = [0] #velocity

t = np.arange(0, 30, 0.1)
h = 0.1 #interval


for i in range(len(t)-1):
  k1 = h*v[i]
  l1 = h*-(x[i])

  k2 = h*(v[i]+l1)
  l2 = h*-(x[i]+k1)

  x_value = x[i]+0.5*(k1+k2)
  x.append(x_value)

  v_value = v[i]+0.5*(l1+l2)
  v.append(v_value)


data_table = pd.DataFrame()
data_table['Time'] = t
data_table['Position'] = x
data_table['Velocity'] = v

print("Data Table For SHO RK 2nd Order")
print(data_table.head(10)) #change here for more values
print(" ")
print(" ")
print(" ")

with plt.style.context('ggplot'):
  plt.figure(dpi=100)
  plt.plot(t, x, 'r', label='Position')
  plt.plot(t, v, 'k', label='Velocity')

  plt.ylim(-2, 2)

  plt.xlabel('Time')
  plt.ylabel('Result')
  plt.title("SHO Runge Kutta Method[2nd Order]", fontsize= '11')
  plt.grid('on')
  plt.legend()
```

#4th order Runge Kutta Method

```python
# initial condition
rk4_x = [1] #position
rk4_v = [0] #velocity

t = np.arange(0, 30, 0.1)
h = 0.1 #interval


for i in range(len(t)-1):
  k1 = h*rk4_v[i]
  l1 = h*-(rk4_x[i])

  k2 = h*(rk4_v[i]+0.5*l1)
  l2 = h*-(rk4_x[i]+0.5*k1)

  k3 = h*(rk4_v[i]+0.5*l2)
  l3 = h*-(rk4_x[i]+0.5*k2)

  k4 = h*(rk4_v[i]+l3)
  l4 = h*-(rk4_x[i]+k3)

  x_value = rk4_x[i]+(1/6)*(k1+2*k2+2*k3+k4)
  rk4_x.append(x_value)

  v_value = rk4_v[i]+(1/6)*(l1+2*l2+2*l3+l4)
  rk4_v.append(v_value)
data_table = pd.DataFrame()
data_table['Time'] = t
data_table['Position'] = rk4_x
data_table['Velocity'] = rk4_v

print("Data Table For SHO RK 4th Order")
print(data_table.head(10))#change here for more values
print(" ")
print(" ")
print(" ")

with plt.style.context('ggplot'):
  plt.figure(dpi=100)
  plt.plot(t, rk4_x, 'r--', label='Position')
  plt.plot(t, rk4_v, 'k--', label='Velocity')

  plt.ylim(-2, 2)

  plt.xlabel('Time')
  plt.ylabel('Result')
  plt.title("SHO Runge Kutta Method[4th Order]", fontsize= '11')
  plt.grid('on')
  plt.legend()
```

```python
#position vs Time 
h = 0.05
nrow, ncol, count = 2,4,1


fig = plt.figure(figsize=(20,10))

with plt.style.context('ggplot'):
    for inv in range(8):
        plt.subplot(nrow, ncol, count)
        x = [1]
        v = [0]

        rk4_x = [1] #position
        rk4_v = [0] #velocity

        t = np.arange(0, 30, 0.1)
        for i in range(len(t)-1):
            k1 = h*v[i]
            l1 = h*-(x[i])
            k2 = h*(v[i]+l1)
            l2 = h*-(x[i]+k1)
            x_value = x[i]+0.5*(k1+k2)
            x.append(x_value)
            v_value = v[i]+0.5*(l1+l2)
            v.append(v_value)

            k1 = h*rk4_v[i]
            l1 = h*-(rk4_x[i])
            k2 = h*(rk4_v[i]+0.5*l1)
            l2 = h*-(rk4_x[i]+0.5*k1)
            k3 = h*(rk4_v[i]+0.5*l2)
            l3 = h*-(rk4_x[i]+0.5*k2)
            k4 = h*(rk4_v[i]+l3)
            l4 = h*-(rk4_x[i]+k3)
            x_value = rk4_x[i]+(1/6)*(k1+2*k2+2*k3+k4)
            rk4_x.append(x_value)
            v_value = rk4_v[i]+(1/6)*(l1+2*l2+2*l3+l4)
            rk4_v.append(v_value)

        plt.plot(t, x,'r', label= "Position vs Time RK2")
        plt.plot(t, rk4_x, 'k--',  label = "Position vs Time RK4")

        plt.ylim(-1.5, 1.5)
        h = round(h,3)
        plt.title("h="+str(h), fontsize= '11')
        plt.grid('on')
        plt.xlabel('Time')
        plt.ylabel('Position')
        plt.legend()
        plt.suptitle('Position VS Time comparison with RK 2nd order and 4th order in various interval', fontweight='bold')

        h = h + 0.05
        count = count+1
```
